```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)  #diagonal element
g.map_upper(plt.scatter)  #upper element
g.map_lower(sns.kdeplot)  #lower element
```
