```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
t = sns.load_dataset('titanic')
```

```python
t.head()
```

```python
tc=t.corr()
```

```python
sns.heatmap(tc,annot=False,cmap='coolwarm')
```

```python
sns.countplot(x='sex',data=t)
```

```python
sns.swarmplot(x='pclass',y='age',data=t)
```

```python
sns.boxplot(y='age',x='pclass',data=t)
```

```python
sns.distplot(t['fare'],kde=False,bins=30)
```

```python
sns.jointplot(x='fare',y='age',data = t,kind='scatter')
```

```python
g = sns.FacetGrid(data = t,col='sex') #important
```

```python
g.map(sns.distplot,'age')
```
